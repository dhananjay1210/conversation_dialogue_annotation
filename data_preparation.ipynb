```python
# function to clean tag name
def clean_tag_name(tag_name):
    tag_name = tag_name.lower()
    tag_name = re.sub('[^a-zA-Z0-9]', ' ', tag_name)
    tag_name = re. sub('\s{2,}', ' ', tag_name)
    tag_name = re. sub('\s', '-', tag_name).strip('-')
    return tag_name
```

```python
annotation_tag_list = []
with open(os.path.join(root_filepath, 'data', 'example_file_for_multigo.tsv'), 'r') as fp:
    for line in fp:
        lst = line.split('\t')
        annotation_tag_list.append(clean_tag_name(lst[0]))
print(annotation_tag_list)
print(len(annotation_tag_list))
```
